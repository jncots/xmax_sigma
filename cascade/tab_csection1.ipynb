```python
import matplotlib.pyplot as plt
from tab_csection import TabulateCSMCEq
tab_cs = TabulateCSMCEq("DPMJETIII191")
```

```python

tab_cs.set_energy_grid(npoints=1000)
sigma = tab_cs.get_sigma()
energy_grid = tab_cs.get_energy_grid()
pid_from_pdg = tab_cs.get_pid_from_pdg()
```

```python
for pdg in pid_from_pdg:
    plt.semilogx(energy_grid, sigma[pid_from_pdg[pdg], :], label=f"{pdg}")
        
plt.grid()
plt.legend()  
```
