```python
%load_ext autoreload
%autoreload 2
import matplotlib.pylab as plt
import numpy as np

from pathlib import Path
import sys
sys.path.insert(0, str(Path("__file__").absolute().parents[1]/"fluka_comparison"))

from fluka_muon_data import fluka_en_data
```

```python
en_spec_dev = fluka_en_data("develop_version")
en_spec_cur = fluka_en_data("current_version")
```

```python
for i, spec_data in enumerate(en_spec_dev):
    spec_data = en_spec_dev[i]
    line = plt.stairs(spec_data[1], spec_data[0], label = f"dev, xdepth={spec_data[3]}")
    spec_data = en_spec_cur[i]
    plt.stairs(spec_data[1], spec_data[0],
               color = line.get_edgecolor(),
               linestyle = "--", 
               label = f"cur, xdepth={spec_data[3]}")  

plt.title(r"Fluka dev vs. current (2021) version, $E_{0}=100$ GeV, $\theta=30^{\circ}$")   
plt.xscale("log")
plt.xlabel("E, GeV")
plt.ylabel(r"$\frac{dN}{dE}\Delta E,$ muons/proton")
# plt.yscale("log")
plt.legend() 
```

```python
for i, spec_data in enumerate(en_spec_dev):
    spec_data = en_spec_dev[i]
    spec_dev = spec_data[1]
    # line = plt.stairs(spec_data[1], spec_data[0], label = f"dev, xdepth={spec_data[3]}")
    spec_data = en_spec_cur[i]
    plt.stairs(spec_dev/spec_data[1], spec_data[0],
            #    color = line.get_edgecolor(),
            #    linestyle = "--", 
               label = f"X={spec_data[3]} g/cm2")  

plt.title(r"Fluka dev/current (2021) version, $E_{0}=100$ GeV, $\theta=30^{\circ}$")       
plt.xscale("log")
plt.xlabel("E, GeV")
plt.ylabel(r"dev/current $\frac{dN}{dE}\Delta E,$")
plt.ylim(0.7, 1.2)
# plt.yscale("log")
plt.legend() 
```
