```python
    def get_ctau(self, pdg):
        """Returns c*tau, in cm"""
        ctau = self.all_particles_dict[pdg].ctau
        if (ctau is None) or math.isinf(ctau):
            return np.inf
        else:
            # 1e-1 is conversion factor from mm to cm
            return np.float64(ctau * 1e-1)
```
