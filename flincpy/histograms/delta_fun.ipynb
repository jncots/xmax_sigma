```python
import numpy as np
```

```python
def ff(r):
    if r <= -2:
        return 0
    elif -2 <= r <= -1:
        return (5 + 2*r-np.sqrt(- 7 - 12*r - 4*r**2))/8
    elif -1 <= r <= 0:
        return (3 + 2*r + np.sqrt(1- 4*r -4*r**2))/8
    elif 0< r < 1:
        return (3-2*r + np.sqrt(1+4*r-4*r**2))/8
    elif 1 <= r <= 2:
        return (5 - 2*r -np.sqrt(-7 + 12*r - 4*r**2))/8
    elif 2 <= r:
        return 0
    else:
        raise ValueError("should not happen")
```

```python
import matplotlib.pylab as plt
```

```python
xgrid = np.linspace(-2, 2, 1000)
myfun = np.frompyfunc(lambda x: np.float64(ff(x)), 1, 1)
myfun_res = myfun(xgrid).astype("float64")
```

```python
import scipy.integrate as integrate
result = integrate.quad(lambda x: ff(x), -10, 10)
result
```

```python
plt.plot(xgrid, myfun_res)
```
