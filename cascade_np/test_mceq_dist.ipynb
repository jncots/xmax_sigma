```python
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
from mceq_comparison import MCEQDistributions
```

```python
mceq_flux = MCEQDistributions(
                 energy = 1e2,
                 pdg_id = 2212,
                 theta_deg = 30,
                 slant_depth = 680,
                 pname_tuples = [
                     ("mu", "mu+", "mu-"),
                     ("numu", "numu", "antinumu"),
                     ("nue", "nue", "antinue"),
                     ("pi", "pi+", "pi-")],
                 interaction_model = "DPMJET-III-19.1", 
                 enable_energy_loss = True, 
                 muon_helicity_dependence = True,
                 disable_decays = [])
```

```python

plt.stairs(mceq_flux.flux["mu"], mceq_flux.e_bins, 
           label = r"${\mu}^{+} + {\mu}^{-}$ mceq", linestyle='--')
plt.stairs(mceq_flux.flux["numu"], mceq_flux.e_bins, 
           label = r"$\bar{\nu}_{\mu} + {\nu}_{\mu}$ mceq", linestyle='--')
plt.stairs(mceq_flux.flux["nue"], mceq_flux.e_bins, 
           label = r"$\bar{\nu}_{e} + {\nu}_{e}$ mceq", linestyle='--')
plt.stairs(mceq_flux.flux["pi"], mceq_flux.e_bins, 
           label = r"$\bar{\pi}^{+} + {\pi}^{-}$ mceq", linestyle='--')

plt.xscale("log")
plt.xlim(1e-2, 2e2)
# plt.ylim(-0.01, 0.31)
plt.legend()
plt.grid()
```
